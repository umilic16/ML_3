## Predobrada podataka i kreiranje train test splita


```python
df_enc = df.copy()
```

```python
cols_to_drop = [
    "arrival_date",
    "arrival_year",
    "no_of_previous_cancellations",
    "no_of_previous_bookings_not_canceled",
    "no_of_week_nights",
    "no_of_weekend_nights",
    "no_of_adults",
    "no_of_children",
    "repeated_guest",
    "required_car_parking_space",
    "no_of_special_requests",
]


df_enc["type_of_meal_plan"].where(
    df_enc["type_of_meal_plan"] == "Meal Plan 1", "other", inplace=True
)
df_enc["room_type_reserved"].where(
    df_enc["room_type_reserved"] == "RT1", "other", inplace=True
)
df_enc["market_segment_type"].where(
    df_enc["market_segment_type"] == "Online", "other", inplace=True
)

df_enc["no_of_people"] = df_enc["no_of_adults"] + df_enc["no_of_children"]
df_enc["no_of_nights"] = df_enc["no_of_week_nights"] + df_enc["no_of_weekend_nights"]

df_enc.drop(
    columns=[
        "no_of_adults",
        "no_of_children",
        "no_of_week_nights",
        "no_of_weekend_nights",
    ]
)
df_enc["has_special_requests"] = df_enc["no_of_special_requests"] > 1.0

df_enc.drop(columns=cols_to_drop, inplace=True)
```

```python
df_enc
```

```python
df_enc["type_of_meal_plan"] = df_enc["type_of_meal_plan"].map(
    {"Meal Plan 1": 0, "other": 1}
)
df_enc["room_type_reserved"] = df_enc["room_type_reserved"].map({"RT1": 0, "other": 1})
df_enc["market_segment_type"] = df_enc["market_segment_type"].map(
    {"Online": 0, "other": 1}
)
df_enc["has_special_requests"] = df_enc["has_special_requests"].astype(int)
```

```python
df_enc
```

```python
X = df_enc.drop("booking_status", axis=1)
y = df_enc["booking_status"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.value_counts()
```

```python
y_train.isna().sum()
```

#### oversampling


```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_bal1, y_train_bal1 = smote.fit_resample(X_train, y_train)
```

```python
y_train_bal1.value_counts()
```

### undersampling


```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_train_bal2, y_train_bal2 = rus.fit_resample(X_train, y_train)
```

```python
y_train_bal2.value_counts()
```

### skaliranje


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_std1 = scaler.fit_transform(X_train_bal1)
X_train_std2 = scaler.fit_transform(X_train_bal2)
```

## Kreiranje modela


```python
X_test_std = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense

model_ovr = Sequential()

model_ovr.add(
    LSTM(units=50, return_sequences=True, input_shape=(X_train_std1.shape[1], 1))
)
model_ovr.add(Dropout(0.2))
model_ovr.add(LSTM(units=50, return_sequences=True))
model_ovr.add(Dropout(0.2))
model_ovr.add(LSTM(units=50))
model_ovr.add(Dropout(0.2))
model_ovr.add(Dense(units=1))

model_ovr.compile(optimizer="adam", loss="mean_squared_error", metrics=["accuracy"])
hist = model_ovr.fit(
    X_train_std1,
    y_train_bal1,
    epochs=25,
    batch_size=32,
    validation_data=(X_test_std, y_test),
    verbose=1,
)
```

```python
plt.subplot(1, 2, 1)
plt.plot(hist.history["loss"])
plt.plot(hist.history["val_loss"])
plt.title("Model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["Training", "Validation"], loc="upper right")
plt.subplot(1, 2, 2)
plt.plot(hist.history["accuracy"])
plt.plot(hist.history["val_accuracy"])
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Training", "Validation"], loc="lower right")
plt.show()
```

```python
model_und = Sequential()

model_und.add(
    LSTM(units=50, return_sequences=True, input_shape=(X_train_std1.shape[1], 1))
)
model_und.add(Dropout(0.2))
model_und.add(LSTM(units=50, return_sequences=True))
model_und.add(Dropout(0.2))
model_und.add(LSTM(units=50))
model_und.add(Dropout(0.2))
model_und.add(Dense(units=1))

model_und.compile(optimizer="adam", loss="mean_squared_error", metrics=["accuracy"])
hist2 = model_und.fit(
    X_train_std2,
    y_train_bal2,
    epochs=25,
    batch_size=32,
    validation_data=(X_test_std, y_test),
    verbose=1,
)
```

```python
plt.subplot(1, 2, 1)
plt.plot(hist2.history["loss"])
plt.plot(hist2.history["val_loss"])
plt.title("Model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["Training", "Validation"], loc="upper right")
plt.subplot(1, 2, 2)
plt.plot(hist2.history["accuracy"])
plt.plot(hist2.history["val_accuracy"])
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Training", "Validation"], loc="lower right")
plt.show()
```
