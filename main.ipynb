### Eksploracija kolona

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    x="no_of_adults", hue="booking_status", data=df
)
plt.show()
```

```python
df['no_of_adults'] = df['no_of_adults'].apply(lambda x: 1 if x == 2 else 0)
df.rename(columns={'no_of_adults': '2_adults_bool'}, inplace=True)

df['2_adults_bool'].value_counts()
```

```python
sns.countplot(
    x="no_of_children", hue="booking_status", data=df
)
plt.show()
```

```python
df['no_of_children'] = df['no_of_children'].apply(lambda x: 0 if x == 0 else 1)
df.rename(columns={'no_of_children': 'any_children_bool'}, inplace=True)

df['any_children_bool'].value_counts()
```

```python
sns.countplot(
    x="no_of_weekend_nights", hue="booking_status", data=df
)
plt.show()
```

```python
df['no_of_weekend_nights'] = df['no_of_weekend_nights'].apply(lambda x: 'More' if x > 2 else x)
df['no_of_weekend_nights'].value_counts()
```

```python
sns.countplot(
    x="no_of_week_nights", hue="booking_status", data=df
)
plt.show()
```
